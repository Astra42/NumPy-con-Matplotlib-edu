```python
def Mus(x, y):
  return y + x
#Construct an array by executing a function over each coordinate.
a = np.fromfunction(Mus, (2, 2))
a = np.fromfunction(lambda x,y: x+y, (2, 3))
a
```

fromiter - создает массив на основе любого итерируемого объекта


```python
np.fromiter("HOLA", dtype="U2")
```

```python
def getSquares(N):
  for i in range(N):
    yield i**2


a = np.fromiter(getSquares(4), dtype="int8")
a
```
