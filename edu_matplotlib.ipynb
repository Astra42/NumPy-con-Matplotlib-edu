```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)

plt.plot(x, y)
plt.fill_between(x, y, 0.5, where=(y<0))#форма между прямой y=0.5 и при условии y<0
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, where=(y<0), color='r', alpha=0.5)
plt.fill_between(x, y, where=(y>0), color='g', alpha=0.5)
plt.grid()
plt.show()
```

##3 Отображение нескольких координатных осей в одном окне 

**subplot(nrows, ncols, index)**



*    nrows, ncols – число строк и столбцов
*   index – индекс текущих координатных осей

image.png





функция subplot  возвращает ссылку на созданную ось (объект axis)

```python
plt.subplot(1,3,1)
plt.plot(np.random.random(10))

plt.subplot(1,3,2)
plt.plot(np.random.random(10))

plt.subplot(1,3,3)
plt.plot(np.random.random(10))

plt.grid()#применятеся к последней активной оси
plt.show()
```

```python
ax1 = plt.subplot(1,3,1)
plt.plot(np.random.random(10))

ax2 = plt.subplot(1,3,2)
plt.plot(np.random.random(10))

ax3 = plt.subplot(1,3,3)
plt.plot(np.random.random(10))

ax1.grid(), ax2.grid(), ax3.grid()#используем принципы ооп
plt.show()
```

Функция **subplots()**

Однако, с помощью функции **subplot()** не очень удобно оперировать большим числом координатных осей в пределах одной формы. Для этого лучше подходит другая похожая функция:

**subplots(nrows, ncols)**

которая на выходе дает ссылку на фигуру и список координатных осей:

```python
f, ax = plt.subplots(2,2)

plt.show()
```

```python
ax[0,0]
```

```python
f, ax = plt.subplots(2,2)


f.set_size_inches(7,4)
f.set_facecolor('#ee5')

ax[0,0].plot(np.arange(0,5,0.2))
ax[0,0].grid()

ax[0,1].plot(np.arange(2,-1,-0.1))
ax[0,1].grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7,4))
plt.plot(np.arange(0,5,0.2)) #ось сразу же применяется к существующей фигуре
plt.show()
```

```python
fig = plt.figure(figsize=(7,4))
ax1 = fig.add_axes([0.0, 0, 1.0, 1.0])#расположение и размер осей(в долях)
ax1.plot(np.arange(0,5,0.2))#укажем эту ось
```

```python
fig = plt.figure(figsize=(7,4))
ax1 = fig.add_subplot(1, 3, 3)#позиционируем в фигуре наж подграфик
ax1.plot(np.arange(0,5,0.2))#укажем эту ось
```

```python
fig = plt.figure(figsize=(7,4))
ax1 = fig.add_subplot(1, 3, 3)
```

Компоновка графиков с помощью **GridSpec**

```python
from matplotlib.gridspec import GridSpec
```

```python
fig = plt.figure(figsize=(7, 4))
gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=[1,2,5], height_ratios=[2, 0.5])#число строк и столбцов и ссылка на фигуру  иии размеры строк и столбцов

ax1 = plt.subplot(gs[0, 0])
ax1.plot(np.arange(0, 5, 0.2))
ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.random(10))
ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.random(10))

plt.show()
```
